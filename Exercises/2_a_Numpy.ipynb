```python
#3.1: import the numpy library
import numpy as np
```

```python
#3.2: create a random 10x10 array named A
A = np.random.rand(10,10)
print(A)
```

```python
#3.3: return only the odd rows of A
odd_rows = list()
odd = np.arange(1, 10, 2)
print(odd)
for i in odd:
    odd_rows.append(A[i])
print(odd_rows)

# easier solution:
odd_rows_easy = A[1:10:2]
print((odd_rows == odd_rows_easy).all()) #True, same result
```

```python
#3.4: perform a matrix multiplication of A with it's transposed
matrix_A = np.matrix(A)
multiplied_A = matrix_A * matrix_A.T
print(multiplied_A)
```

```python
#3.5: compute the matrix inverse of A
inverse_A = matrix_A.I
print(inverse_A)
```

```python
#3.6: divide all entries of A which are lager than 0.5 by 2 (hint: use where() )
```

```python
B = np.copy(A)
print("Before: ")
print(B)
mask = (B > 0.5)
print("Entries larger than 0.5: ")
print(B[mask])
B[mask] = B[mask]/2
print("Entries larger than 0.5, devided by 2: ")
print(B[mask])
print("After: ")
print(B)
```
