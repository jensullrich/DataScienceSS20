## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
plt.plot(data.month_number, data.total_profit)
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Total profit of all months')
plt.show()
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
fig, ax = plt.subplots()

ax.plot(data.month_number, data.facecream, color="red")
ax.plot(data.month_number, data.toothpaste, color="blue")       
ax.plot(data.month_number, data.facewash, color="green")
ax.plot(data.month_number, data.bathingsoap, color="purple")
ax.plot(data.month_number, data.shampoo, color="cyan")
ax.plot(data.month_number, data.moisturizer, color="magenta")
ax.legend()
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.show()
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
plt.scatter(data.month_number, data.toothpaste)
plt.grid(True)
plt.show()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
fig, ax = plt.subplots()
width = 0.4
ax.bar(data.month_number-width/2, data.facecream, width, label="facecream")
ax.bar(data.month_number+width/2, data.facewash, width, label="facewash")
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
ax.legend()
plt.show()
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 

```python
toothpaste = data.toothpaste.sum()
facewash = data.facewash.sum()
bathingsoap = data.bathingsoap.sum()
facecream = data.facecream.sum()
shampoo = data.shampoo.sum()
moisturizer = data.moisturizer.sum()

fig, ax = plt.subplots()
labels = 'toothpaste', 'facewash', 'bathingsoap', 'facecream', 'shampoo', 'moisturizer'
sizes = [toothpaste, facewash, bathingsoap, facecream, shampoo, moisturizer]
ax.pie([toothpaste, facewash, bathingsoap, facecream, shampoo, moisturizer], labels=labels, autopct='%1.1f%%')
ax.axis('equal')
plt.show()
```
